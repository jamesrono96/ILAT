```python
# -*- coding: utf-8 -*-
"""Simple Library Management System.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1CyIwFC1Ov-V7jkN_UAZ8MEbl2MoHV7-W
"""

!pip install colorama

# Installing colorama for colored output
!pip install colorama

# Import necessary libraries
from colorama import init, Fore, Style
init(autoreset=True)  # Initialize colorama and auto-reset styles

# List to store book records
books = []

# Function to create a colorful box around text
def print_boxed(text, border_color=Fore.CYAN, text_color=Fore.WHITE):
    border_line = border_color + "+" + "-" * (len(text) + 4) + "+"
    print(border_line)
    print(border_color + "|" + text_color + " " * (len(text) + 2) + "|")
    print(border_color + "|" + text_color + " " + text + " " + "|")
    print(border_color + "|" + text_color + " " * (len(text) + 2) + "|")
    print(border_line)

# Function to add a book
def add_book():
    print_boxed("Add Book", Fore.CYAN, Fore.WHITE)
    title = input(Fore.GREEN + "Enter book title: ")
    author = input(Fore.GREEN + "Enter book author: ")
    isbn = input(Fore.GREEN + "Enter book ISBN: ")
    new_book = {'title': title, 'author': author, 'isbn': isbn, 'available': True}
    books.append(new_book)
    print_boxed("Book added successfully.", Fore.YELLOW, Fore.BLACK)

# Function to search for a book
def search_book():
    print_boxed("Search Book", Fore.CYAN, Fore.WHITE)
    keyword = input(Fore.GREEN + "Enter title or author to search: ")
    results = [book for book in books if keyword.lower() in book['title'].lower() or keyword.lower() in book['author'].lower()]
    if results:
        print_boxed("Search Results:", Fore.YELLOW, Fore.BLACK)
        for book in results:
            print(Fore.MAGENTA + f"Title: {book['title']}, Author: {book['author']}, ISBN: {book['isbn']}, Available: {book['available']}")
    else:
        print_boxed("No books found.", Fore.RED, Fore.WHITE)

# Function to checkout a book
def checkout_book():
    print_boxed("Checkout Book", Fore.CYAN, Fore.WHITE)
    isbn = input(Fore.GREEN + "Enter ISBN of the book to checkout: ")
    member_id = input(Fore.GREEN + "Enter member ID: ")
    for book in books:
        if book['isbn'] == isbn:
            if book['available']:
                book['available'] = False
                book['checked_out_by'] = member_id
                print_boxed("Book checked out successfully.", Fore.YELLOW, Fore.BLACK)
                return
            else:
                print_boxed("Book is currently unavailable.", Fore.RED, Fore.WHITE)
                return
    print_boxed("Book not found.", Fore.RED, Fore.WHITE)

# Function to return a book
def return_book():
    print_boxed("Return Book", Fore.CYAN, Fore.WHITE)
    isbn = input(Fore.GREEN + "Enter ISBN of the book to return: ")
    member_id = input(Fore.GREEN + "Enter member ID: ")
    for book in books:
        if book['isbn'] == isbn:
            if not book['available'] and book.get('checked_out_by') == member_id:
                book['available'] = True
                del book['checked_out_by']
                print_boxed("Book returned successfully.", Fore.YELLOW, Fore.BLACK)
                return
            else:
                print_boxed("Invalid return.", Fore.RED, Fore.WHITE)
                return
    print_boxed("Book not found.", Fore.RED, Fore.WHITE)

# Main function to interact with the system
def main():
    while True:
        print_boxed("Library Management System", Fore.CYAN, Fore.WHITE)
        print(Fore.GREEN + "1. Add Book")
        print(Fore.GREEN + "2. Search Book")
        print(Fore.GREEN + "3. Checkout Book")
        print(Fore.GREEN + "4. Return Book")
        print(Fore.GREEN + "5. Exit")

        choice = input(Fore.YELLOW + "Enter your choice (1-5): ")

        if choice == '1':
            add_book()
        elif choice == '2':
            search_book()
        elif choice == '3':
            checkout_book()
        elif choice == '4':
            return_book()
        elif choice == '5':
            print_boxed("Exiting the system.", Fore.GREEN, Fore.BLACK)
            break
        else:
            print_boxed("Invalid choice. Please select a number between 1 and 5.", Fore.RED, Fore.WHITE)

# Run the main function
main()
```
